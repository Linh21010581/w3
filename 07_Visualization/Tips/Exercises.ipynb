### Step 1. Import the necessary libraries:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
tips = pd.read_csv(url)
```

### Step 3. Assign it to a variable called tips

```python
print(tips.head())
```

### Step 4. Delete the Unnamed 0 column

```python
if 'Unnamed: 0' in tips.columns:
    tips.drop(columns=['Unnamed: 0'], inplace=True)
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(8, 5))
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Histogram of Total Bill")
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip', size='size', hue='size', palette='coolwarm', alpha=0.7)
plt.title("Relationship of Total Bill, Tip, and Size")
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title("Total Bill Distribution by Day")
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='tip', y='day', hue='sex')
plt.title("Tip vs Day Differentiated by Gender")
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips, x='day', y='total_bill', hue='time')
plt.title("Total Bill by Day and Time")
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(tips[tips['time'] == 'Dinner']['tip'], bins=15, kde=True, ax=axes[0], color='blue')
axes[0].set_title("Tip Distribution (Dinner)")
sns.histplot(tips[tips['time'] == 'Lunch']['tip'], bins=15, kde=True, ax=axes[1], color='red')
axes[1].set_title("Tip Distribution (Lunch)")
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.scatterplot(data=tips[tips['sex'] == 'Male'], x='total_bill', y='tip', hue='smoker', ax=axes[0])
axes[0].set_title("Total Bill vs Tip (Male)")

sns.scatterplot(data=tips[tips['sex'] == 'Female'], x='total_bill', y='tip', hue='smoker', ax=axes[1])
axes[1].set_title("Total Bill vs Tip (Female)")

plt.show()
```
